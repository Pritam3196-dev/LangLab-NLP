### Word2Vec

#### Word2vec is feature extraction technique like bow and Tfidf but it is differ from it . it gives sementic meaning of words . If those words are in similar context then place those words close to each other in hight dimensional space
#### it has two types
#### 1.CBOW (Continous bag of words)
#### 2.Skipgram



##### 1.CBOW : Basically it predicts center or target word on basis of surrounding context word
##### 2.skipgram : skipgram takes cnter or target word on basis of it . it make prediction of surrounding words
