### Tokenization , POS Tagging , Dependency parsing

#### Here when we apply for loop so loop iterate object to object means character to character but we put data into spacy_model so already make into word tokenize. so iteration will peroform on word by word

##### From word we will extract its text and pos_tagging and dependency parsing means it tells grammatical structure of sentence , relation between words   

```python
len(doc)
```

```python

for word in doc:
    
    print('word/ Token', '---->' , word.text ,    '      Pos_tag','---->' ,word.pos_ , 
          '      dep parse' , '------>' , word.dep_)
```

### Named Entity Recognition

```python
doc1 = spacy_model("Elon Musk, the CEO of SpaceX, recently announced a new project. SpaceX's headquarters is located in Hawthorne, California. The project aims to launch rockets to Mars by 2025. Musk has also expressed his interest in electric vehicles through his work with Tesla Inc.")
```

```python
print(doc1)
```

#### from text extracts names and label_ idntify/ recognise name of places and peoples

```python
for name in doc1.ents:    # ents is entities
    print('Name','--->',name.text,'     Label Of Name','----->',name.label_)
```

### As I mentioned earlier it is small size light weight fast but less accurate as compare to larger models so thats why it show spacex is person and GPE than organisation

### Lemmatization

#### Pos_tag means identify grammatical role of word and lemmatization reduce word in meaningful base word

```python
for word in doc1:
    
    print( 'Token/word','----->' ,word.text, '    Pos_tag','---->',word.pos_,
          '        Lemmatization','------>',word.lemma_)
```

#### Visualization by using spacy

```python
from IPython.core.display import display ,HTML
from spacy import displacy
import warnings                      # this two just for avoid unnesessary warnings
warnings.filterwarnings('ignore')
```

##### display is function from Ipython library its actually to allow various type of data (eg:text) in jupyter notebook . it is used when jupyter notebook render output in readable format
##### HTML is class it is used to render HTML content in jupyter notebook

#### Another one visualization on NER

### stopwords

#### Extract stopwords from spacy model

```python
stopwords = spacy_model.Defaults.stop_words
print(stopwords)
```

### We want words from our document without stopwords

```python
token_text =[]

for token in doc1:
    
    token_text.append(token)
    
#print(token_text)
tokens_without_sw = [word.text for word in token_text if word.text not in stopwords]
print(tokens_without_sw)
```

#### word.text means extract text from iterating variable(token)
